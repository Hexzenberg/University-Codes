Case Study: Library Management System

Problem Statement:

You're tasked with creating a library management system where users can borrow and return books. Each book has attributes like title, author, and availability status.


The Book class represents a book with attributes like title, author, and availability. It has methods like borrow() and return_book() to change the availability status.

The Library class represents a library, which has a list of books. It has methods like add_book() to add books to the library and display_available_books() to show available books.

We create instances of the Book class and add them to the Library. Then, we demonstrate borrowing and returning books.

This case study demonstrates the use of functions (methods) within classes to perform specific actions and encapsulation of data within objects, which are fundamental concepts in both functions and OOP.

```python
#book class
class book:
    
    #initialising constructor
    def __init__(self, title, author, availability):
        self.title = title
        self.author = author
        self.availability = availability
    
    #borrow function returns 1 if book is available and 0 when it's not 
    def borrow(self):
        if self.availability > 0:
            self.availability -= 1
            return 1
        else:
            return 0
        
    #return_book function adds 1 to the availability of that book
    def return_book(self):
        self.availability += 1

#library class
class library:
    
    def __init__(self):
        self.books = book[]
    
    #add_book function adds a book in the library
    def add_book(self, book):
        self.books.append(book)
        
    #finds book if it's available in the library or not
    def find_book(self, book):
        for i in self.books:
            if i.title == book and i.availability > 0: return i
        return False
        
    #display_available_books displays all the available book's title in the library
    def display_available_books(self):
        for book in self.books:
            print(book.title, " availability = ", book.availability)
            
    #updating return status
    def returned_book(self, book):
        for i in self.books:
            if i.title == book:
                i.return_book()
                
    #updating borrowed status
    def borrowed_book(self, book):
        for i in self.books:
            if i.title == book:
                i.borrow()
            
my_library = library()

my_library.add_book(book('math', 'abc', 2))
my_library.add_book(book('science', 'abc', 2))
my_library.add_book(book('sst', 'abc', 2))
my_library.add_book(book('hindi', 'abc', 2))


while(1):
    option = int(input("Enter\n1 to display all avalilable books\n2 to borrow a book\n3 to return a book\n4 to exit: "))
    
    if(option == 2 or option == 3):
        book = input("Enter the book name: ")
        my_book = my_library.find_book(book)
    
    
    match option:
        case 1:
            print("\n")
            my_library.display_available_books()
        case 2:
            if my_book:
                print("Book available, go get it!")
                my_library.borrowed_book(book)
            else:
                print("Book not available!üòî")
        case 3:
            my_book.return_book()
            my_library.returned_book(book)

            print("Book returned!")
        case 4:
            print("Come again!")
            break
        case _:
            print("Invalid option!")
```

Experiment: Financial Portfolio Management System

Problem Statement:
Create a program to manage a financial portfolio. This portfolio can contain different types of assets such as stocks, bonds, and cash. Implement functionalities to calculate the total value of the portfolio, add or remove assets, and display detailed information about each asset.


ÔÅ¨	The Asset class represents a financial asset with attributes like name, quantity, and price.

ÔÅ¨	It has a method value() to calculate the total value of that asset.

ÔÅ¨	The Portfolio class represents a financial portfolio, which contains a list of assets.

ÔÅ¨	It has methods like add_asset() to add assets to the portfolio, remove_asset() to remove assets, total_value() to calculate the total value of the portfolio, and display_assets() to display all assets in the portfolio.

ÔÅ¨	We create instances of the Asset class representing different assets and add them to the Portfolio.

ÔÅ¨	We demonstrate adding, removing assets, and calculating the total value of the portfolio.

```python
# Declaring a class for the asset
class Asset:
    def __init__(self, name, quantity, price):
        self.name = name
        self.quantity = quantity
        self.price = price

    def value(self):
        return self.quantity * self.price

# Declaring a class for the portfolio
class Portfolio:
    def __init__(self):
        self.assets = []

    def add_asset(self, asset):
        self.assets.append(asset)

    def remove_asset(self, asset_name):
        for asset in self.assets:
            if asset.name == asset_name:
                self.assets.remove(asset)
                print(f"{asset_name} removed from the portfolio.")
                return
        print(f"{asset_name} not found in the portfolio.")

    def total_value(self):
        total_value = sum(asset.value() for asset in self.assets)
        return total_value

    def display_assets(self):
        if not self.assets:
            print("No assets in the portfolio.")
            return
        print("Assets in the portfolio:")
        for asset in self.assets:
            print(f"{asset.name}: Quantity - {asset.quantity}, Price - ${asset.price}, Total Value - ${asset.value()}")


# Creating instances of assets
stock1 = Asset("AAPL", 10, 150)
bond1 = Asset("US Treasury Bond", 5, 1000)
cash = Asset("Cash", 10000, 1)

# Creating a portfolio
portfolio = Portfolio()

# Adding assets to the portfolio
portfolio.add_asset(stock1)
portfolio.add_asset(bond1)
portfolio.add_asset(cash)

# Displaying assets in the portfolio
portfolio.display_assets()

# Calculating total value of the portfolio
total_value = portfolio.total_value()
print("Total value of the portfolio:", total_value)

# Removing an asset from the portfolio
portfolio.remove_asset("AAPL")

# Displaying assets in the portfolio after removal
portfolio.display_assets()
```
